# IIR low-pass filter demo 1

```python
from __future__ import print_function
import numpy as np
import scipy.signal as signal
from ipywidgets import interact, interactive, fixed, interact_manual
from signal_plot import signal_plot
```

A first-order recursive filter has a difference equation:

$y[n] = \alpha y[n - 1]  + (1 - \alpha) x[n]$

where $\alpha$ is a parameter that controls the break frequency.  When $\alpha = 0$ there is no filtering.  The closer $\alpha$ gets to 1, the greater the filtering.

```python
A = 1
N = 100
fs = 100
sigma = 0.2
mu = 0
np.random.seed(42)
n = np.random.standard_normal(N) * sigma + mu
t = np.arange(N) / fs

def iir_lpf_filter_plot(f=4, alpha=0.5):
    s = A*np.sin(2*np.pi*f*t)
    x = s + n

    y = signal.lfilter(b=(1 - alpha, ), a=(1, -alpha), x=x)

    signal_plot(t, y, both=True)    

interact(iir_lpf_filter_plot, alpha=(0.0, 1.0, 0.01))
```
